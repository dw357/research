```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import cm
from skimage import measure

import os,sys
from PIL import Image
%matplotlib inline
```

```python
# Input
#     x: x-cooridinate of ring center
#     y: y-coordinate of ring center 
#     di: inner diameter of ring in nm
#     do: outer diameter of ring in nm

def draw_ring(x, y, di, do):
    ri = di/2.0
    ro = do/2.0
    ang = np.linspace(0, 2*np.pi, num = 500)
    xi = ri * np.cos(ang)
    yi = ri * np.sin(ang) 
    xo = ro * np.cos(ang)
    yo = ro * np.sin(ang) 
    ax2.plot(x+xi, y+yi, 'w')
    ax2.plot(x+xo, y+yo, 'w')
    return 
```

```python
# Draws two rectangles on ax4, (1000 x 200)nm and (500 x 200) nm
# Assumes data file is 1050 x 300 in pixels 
# Input
#     x_c: x-cooridinate of bottom left corner of rectangle
#     y_c: y-coordinate of bottom left corner of rectangle
#     di: width in nm
#     do: height in nm
#     x_file: number of pixels in the x-direction (from data file, check Lumerical Edit Setting for this number) 
#     y_file: number of pixels in the x-direction (from data file, check Lumerical Edit Setting for this number)

def draw_rec(x_b, y_b, px_x, px_y, x_file, y_file):
    ax4.add_patch(patches.Rectangle(
        (1.0*x_b/x_file*px_x, 1.0*y_b/y_file*px_y),   #(x, y) bottom left corner 
        1000./x_file*px_x,    #width
        200./y_file*px_y,      #height
        fill=False,  color='w'))
    ax4.add_patch(patches.Rectangle(
        (1.0*(x_b+250)/x_file*px_x, 1.0*y_b/y_file*px_y),   #(x, y) bottom left corner 
        500./x_file*px_x,    #width
        200./y_file*px_y,      #height
        fill=False, color='w'))
```

# Without SiO2

```python
spec = np.loadtxt('500_spectrum.txt', delimiter=',')
plt.plot(spec[:,0], spec[:,1])
plt.xlabel('Wavelength (nm)')
plt.title('Emission spectrum of Di = 500 nm, without SiO2')
plt.show()
```

### `402 nm` (First Order) Q = 150,684

```python
data_h = '500_402_h.txt'
data_v = '500_402_v.txt'

### Horizontal H field 
image_h = np.loadtxt(data_h)
px_h = image_h.shape[0]
### Vertical H field
image_v = np.transpose(np.loadtxt(data_v))
px_vx = image_v.shape[1]
px_vy = image_v.shape[0]

# Plot 
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2,2, figsize=(10,8))
ax1.imshow(image_h) 
ax2.imshow(image_h)
ax3.imshow(image_v) 
ax4.imshow(image_v)
draw_ring(px_h/2., px_h/2., 500/1050.*px_h, 1000/1050.*px_h)
draw_rec(25, 50, px_vx, px_vy, 1050, 300)
plt.show()
```
